### PromptTemplate

The prompt object formats my variables, passes the resulting string to the model, and returns the model's response.

````python
from langchain_core.prompts import PromptTemplate

# LangChain’s PromptTemplate stores the reusable instructions separately 
# from the value that changes. Its placeholders use braces such 
# as {text} and are filled at runtime.

# This is an unfinished prompt, just an empty shell/template
# This line below replaces the f string we wrote in the first tutorial 
# of: f"Summarize the following text in one sentence:\n\n{summary_text}"
summary_template = PromptTemplate.from_template(

"""
Summarize the following text in one sentence:

{text}
"""
)

print(summary_template)
````

**Fill the template**

````python
summary_text = (
    "Agentic AI systems combine language models "
    "with tools, memory, and control logic."
)

# formatted_prompt fills the prompt template from above with text
formatted_prompt = summary_template.format(
    text=summary_text
)

print(formatted_prompt)
print("\nType:", type(formatted_prompt))
````

## Call Llama

```
summary_text
    ↓
PromptTemplate.format()
    ↓
formatted_prompt
    ↓
llm.create_chat_completion()
    ↓
summary_result

```

````python
response = llm.create_chat_completion(
    messages=[
        {
            "role": "user",
            "content": formatted_prompt,
        }
    ],
    temperature=0.2,
    max_tokens=100,
)

summary_result = response["choices"][0]["message"]["content"]

print(summary_result)
````
